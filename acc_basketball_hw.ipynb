```python
#part 1 reading the csv
import pandas as pd

acc_players = pd.read_csv('/Users/josh/Downloads/acc_players-2324F.csv', skiprows = 1)

print("ACC Players Stats:")
display(acc_players.head())
```

```python
#part 2 basic analysis

acc_players['PTS'] = pd.to_numeric(acc_players['PTS']) #data cleaned as pts was a string value
total_points = acc_players['PTS'].sum()

print("Total points scored by all players combined:")
total_points
```

```python
most_minutes_player = acc_players[['Player', 'MP']].sort_values(by='MP', ascending=False).head(1)

print("Player with the most minutes played:")
most_minutes_player
```

```python
top_5_rebounders = acc_players[['Player', 'TRB']].sort_values(by='TRB', ascending=False).head(5)

print("Top 5 rebounders:")
top_5_rebounders
```

```python
#part 3 player filtering

players_above_500_minutes = acc_players[acc_players['MP'] > 500]

print("Players with above 500 minutes:")
players_above_500_minutes
```

```python
highest_assists_player = players_above_500_minutes[['Player', 'AST']].sort_values(by='AST', ascending=False).head(1)

print("Player with the highest assists:")
highest_assists_player
```

```python
top_3_assist_leaders = players_above_500_minutes[['Player', 'AST']].sort_values(by='AST', ascending=False).head(3)

print("Top 3 assist leaders:")
top_3_assist_leaders
```

```python
top_3_shot_blockers = players_above_500_minutes[['Player', 'BLK']].sort_values(by='BLK', ascending=False).head(3)

print("Top 3 shot blockers:")
top_3_shot_blockers
```

```python
#part 4 school based analysis
school_points = acc_players.groupby('School')['PTS'].sum().reset_index()

print("Total points scored by each school:")
school_points
```

```python
school_assists = acc_players.groupby('School')['AST'].sum().reset_index()

print("Total assists per team:")
school_assists
```

```python
top_3_schools_by_points = school_points.sort_values(by='PTS', ascending=False).head(3)

print("Top 3 schools with most points scored:")
top_3_schools_by_points
```
